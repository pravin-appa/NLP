```python
# Importing data
import pandas as pd

data = pd.read_csv('SMSSpamCollection',sep='\t', names=["label","message"])
```

```python
data
```

```python
#Data Cleaning and Preprocessing

import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

ps = PorterStemmer()
```

```python
#regular expression

import re

corpus =[]

for i in range(0,len(data)):
  rev = re.sub('[^a-zA-Z0-9]',' ',data['message'][i])
  rev = rev.lower()
  rev = rev.split()
  rev = [ps.stem(word) for word in rev if word not in set(stopwords.words('english')) ]
  rev = " ".join(rev)
  corpus.append(rev)
```

```python
corpus
```

# Create Bag of Words

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=3000, binary=True, ngram_range=(1,2))
```

```python
X = cv.fit_transform(corpus).toarray()
```

```python
X
```

```python
y = pd.get_dummies(data['label'], dtype=int)
y = y.iloc[:,1].values
```

```python
y
```

```python
# Train Test Split

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

```python
X_train,y_train
```

```python
from sklearn.naive_bayes import MultinomialNB
spam_class_model = MultinomialNB()
```

```python
spam_class_model = spam_class_model.fit(X_train, y_train)
```

```python
y_pred = spam_class_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report

score = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
```

```python
print(score), print(report)
```

```python
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('wordnet')
lemmatizer = WordNetLemmatizer()
```

```python
import re
corpus =[]

for i in range(0,len(data)):
    review = re.sub('[^a-zA-Z]', ' ', data['message'][i])
    review = review.lower()
    review = review.split()

    review = [lemmatizer.lemmatize(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
from nltk.tokenize import sent_tokenize
from gensim.utils import simple_preprocess
nltk.download('punkt')
```

```python
words =[]

for sent in corpus:
  sent_token = sent_tokenize(sent)
  for sent in sent_token:
    words.append(simple_preprocess(sent))
```

```python
words
```

```python
# traing word2vec
import gensim
from gensim.models import Word2Vec
```

```python
model = Word2Vec(words,window=5,min_count=2)
```

## Average Word2Vec

```python
import numpy as np
```

```python
def avg_word2vec(doc):
    # remove out-of-vocabulary words
    #sent = [word for word in doc if word in model.wv.index_to_key]
    #print(sent)

    return np.mean([model.wv[word] for word in doc if word in model.wv.index_to_key],axis=0)
                #or [np.zeros(len(model.wv.index_to_key))], axis=0)
```
